# Ans:

「File I/O」:讀取的資料為未處理過的資料，所以是原始的一大串程式碼

「xmltodict」:由於經由module處理過，有將特定資料呈現出來

## File I/O


```python
import xmltodict
with open ("../day3_data/64_72hr_CH.xml", "r") as file:
    data = dict(xmltodict.parse(file.read()))
```

```python
# 1. 請問高雄市有多少地區有溫度資料？
locationName = data['cwbopendata']['dataset']['locations']['location']
print(len(locationName),"區")
region_list = []
for count in range(len(locationName)):
#     print(locationName[count]['locationName'])
    region_list.append(locationName[count]['locationName'])
print("區域名稱:",region_list)
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
locationName = data['cwbopendata']['dataset']['locations']['location']
for count in range(len(locationName)):
    region = locationName[count]['locationName']
    time = locationName[count]['weatherElement'][1]['time'][0]['dataTime']
    temperture = locationName[count]['weatherElement'][1]['time'][0]['elementValue']['value']
    print("%s之第一個時間:%s溫度:攝氏%s"%(region,time,temperture))
#     print(type(temperture))
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
# locationName[0]['weatherElement'][0] #溫度
time_period = len(locationName[0]['weatherElement'][0]['time'])#溫度
print("第一個地區為:%s" %locationName[0]['locationName'])
for num in range(time_period):
    print("時間為:%s" %locationName[0]['weatherElement'][0]['time'][num]['dataTime'])
    print("溫度為:攝氏%s" %locationName[0]['weatherElement'][0]['time'][num]['elementValue']['value'])
    
```
