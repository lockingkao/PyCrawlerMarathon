### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
import requests
content = requests.get("https://api.github.com/events")
content.s
```

```python
import json 
json.loads(content.text)
```

由上面兩個預覽可以發現
json.loads(content.text)傳回的資料有排版
